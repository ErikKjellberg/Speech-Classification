<h3>Viewing attention weights</h3>
<p>In this notebook, a visualization of the attention weights for the HAN model can be done.
In order to view more speeches than the ones provided, one can run the visualize_attention.py in the src directory.
Note that this requires a trained model.</p>
<p>Inspiration for this notebook came from 
<a href="https://github.com/JoungheeKim/Pytorch-Hierarchical-Attention-Network/blob/master/visualize_tutorial.ipynb">this notebook</a> by Kim Jounghee.

```python
import json
from IPython.display import Markdown, display
import html
```

```python
def make_weighted_document(str_list, sent_weights, word_weights):
    temp_str = []
    max_weight = 0
    for i, s_w in enumerate(sent_weights):
        for word_weight in word_weights[i]:
            max_weight = max(max_weight,s_w*word_weight)        
    for i, sent in enumerate(str_list):
        sent_weight = sent_weights[i]
        # Remove CLS and SEP tokens
        for word, word_weight in zip(sent[1:-1],word_weights[i][1:-1]):
            temp_str += ['<span style="background-color:rgba(255,0,255,' + str(word_weight*sent_weight/max_weight) + ');  font-size: 12pt;">' + html.escape(word) + '</span>']
    joined = " ".join(temp_str)
    return joined
```

```python
# Uncomment to visualize the model trained on documents where party names have not been filtered out
#with open("weighted_documents.json", "r", encoding="utf-8-sig") as f:
#    weighted_documents = json.load(f)

with open("weighted_documents_filtered.json", "r", encoding="utf-8-sig") as f:
    weighted_documents = json.load(f)
```

```python
# Choose which documents to view
start, stop = 13,23
for i in range(start,stop):
    sentences = weighted_documents["sentences"][i]
    sent_weights = weighted_documents["sent_weights"][i]
    words = weighted_documents["tok_sentences"][i]
    word_weights = weighted_documents["word_weights"][i]
    pred = weighted_documents["pred"][i]
    true = weighted_documents["true"][i]
    display(Markdown(make_weighted_document(words,sent_weights,word_weights)))
    display(Markdown('<span style="font-size: 12pt;">Predicted: '+str(pred)+'<br>True: '+str(true)+'</span>'))
```
